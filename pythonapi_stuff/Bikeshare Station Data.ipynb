```python
json = requests.get('http://www.bayareabikeshare.com/stations/json')
data = json.json()['stationBeanList']

lon = []
lat = []

def sf():
    sf_stations = filter(lambda x: x['city'] == 'San Francisco', data)
    for i in sf_stations:
        lon.append(i['longitude'])
        lat.append(i['latitude'])
```

```python
m = Basemap(projection='merc', 
            lat_0=37.428, 
            lon_0=-122.145, 
            resolution='h',
            llcrnrlon= -122.520557,
            llcrnrlat= 37.697017,
            urcrnrlon= -122.364448, 
            urcrnrlat= 37.833525)
```

```python
x1, y1 = m(lon, lat)
m.drawmapboundary(fill_color='#90C0F8') # fill to edge
m.drawcountries()
m.fillcontinents(color='white', lake_color='#90C0F8', zorder=0)
m.scatter(x1, y1, marker="o", alpha=1)
plt.title("SF Bikeshare Locations")
plt.show()
```
