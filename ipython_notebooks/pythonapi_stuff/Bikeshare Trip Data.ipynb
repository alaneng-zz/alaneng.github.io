```python
import pandas as pd
import datetime as dt
import numpy as np
```

```python
raw = pd.read_csv('/Users/alaneng/Downloads/babs_open_data_year_2/201508_trip_data.csv')
raw_stn = pd.read_csv('/Users/alaneng/Downloads/babs_open_data_year_2/201508_station_data.csv')
```

```python
data = raw.copy()
data = data.merge(raw_stn, left_on='Start Station', right_on='name')
```

```python
data['Start Date'] = map(lambda x: dt.datetime.strptime(x, "%m/%d/%Y %H:%M"), data['Start Date'])
data['End Date'] = map(lambda x: dt.datetime.strptime(x, "%m/%d/%Y %H:%M"), data['End Date'])
```

## What percentage of trips are by one-off customers?

```python
# customer trips / all trips
round(data[data['Subscriber Type'] == 'Customer'].shape[0] / float(data.shape[0]), 2) * 100
```

## What stations are most heavily used by subscribers in the morning commute in SF?

```python
# Define hours

def commute(x):
    if x.hour >=5 and x.hour < 9:
        return 'Morning'
    elif x.hour >= 9 and x.hour < 17:
        return 'Afternoon'
    elif x.hour >= 17 and x.hour < 21:
        return 'Evening'
    else:
        return 'After Dark'

data['Commute'] = data['Start Date'].apply(lambda x: commute(x))
```

```python
# Where those riding in SF begin their morning ride

data[(data['landmark'] == 'San Francisco') 
     & (data['Commute'] == 'Morning')
     & (data['Subscriber Type'] == 'Subscriber')]['Start Station'].value_counts().head(10)
```

## What start and end point has the longest duration on average in the morning?

```python
x = data[(data['landmark'] == 'San Francisco') 
      & (data['Commute'] == 'Morning')
      & (data['Subscriber Type'] == 'Subscriber')]
```

```python
# Longest duration on average by starting point
(x.groupby(['Start Station'])['Duration'].mean().sort_values(ascending=False) / 60).head(10)
```

```python
# Longest duration on average by starting and end point
pd.DataFrame(x.groupby(['Start Station', 'End Station']).mean()['Duration'].sort_values(ascending=False) / 60).head(10)
```

## Result above shows average duration of 2 hours from Market at Sansome to Market at Sansome. Why?


```python
# Because someone rented the bike for 8 hours
x[(x['Start Station'] == 'Market at Sansome') & (x['End Station'] == 'Market at Sansome')]['Duration']
```

## Use median to return trips that are more indicative of avg time

## Return multiple statistics

```python
(x.groupby(['Start Station', 'End Station'])['Duration'].agg([np.median, np.mean]).sort_values('median', ascending=False) / 60).loc['Steuart at Market', 'Harry Bridges Plaza (Ferry Building)']
```
